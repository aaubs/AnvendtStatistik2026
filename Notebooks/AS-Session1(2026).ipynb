## 4. BEREGN STATISTIK

````python
# Lad os se p√• engagement rates

print("=== ENGAGEMENT RATE STATISTIK ===")
print(f"Gennemsnit: {df['engagement_rate'].mean():.2f}%")
print(f"Median: {df['engagement_rate'].median():.2f}%")
print(f"Minimum: {df['engagement_rate'].min():.2f}%")
print(f"Maximum: {df['engagement_rate'].max():.2f}%")
print(f"Standardafvigelse: {df['engagement_rate'].std():.2f}%")
````

````python
# Eksempel: Hvorfor "print()" er vigtigt
df['engagement_rate'].mean()
df['engagement_rate'].median()
df['engagement_rate'].min()
df['engagement_rate'].max()
df['engagement_rate'].std()
````

````python
# Sammenlign sm√• vs store influencers

# Definer kategorier baseret p√• f√∏lgere
df['size_category'] = pd.cut(df['followers'], 
                               bins=[0, 10000, 100000, 1000000],
                               labels=['Nano (1K-10K)', 'Micro (10K-100K)', 'Macro (100K+)'])

# Beregn gennemsnitlig engagement per kategori
avg_by_size = df.groupby('size_category')['engagement_rate'].mean()

print("=== GENNEMSNITLIG ENGAGEMENT RATE PER ST√òRRELSE ===")
print(avg_by_size)
````

### üîç Interessant M√∏nster!

Kan I se det? 

**Jo flere f√∏lgere, jo LAVERE engagement rate!**

Dette er modsat hvad mange tror. Hvorfor mon det er s√•dan?

- Sm√• influencers: Mere personlige relationer
- Store influencers: Mange "d√∏de" f√∏lgere (bots, inaktive)
- Niche audiences: Mere engagerede end brede audiences

**Dette er statistisk t√¶nkning** - vi finder m√∏nstre der ikke er oplagte.

## 5. VISUALISER M√òNSTRE

````python
# Lav et scatter plot: f√∏lgere vs engagement

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='followers', y='engagement_rate', 
                hue='category', s=100, alpha=0.7)

plt.title('Relationship: Followers vs Engagement Rate', fontsize=16, fontweight='bold')
plt.xlabel('Antal F√∏lgere', fontsize=12)
plt.ylabel('Engagement Rate (%)', fontsize=12)
plt.legend(title='Kategori', bbox_to_anchor=(1.05, 1), loc='upper left')

# Tilf√∏j en trend linje
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(df['followers'], df['engagement_rate'])
line = slope * df['followers'] + intercept
plt.plot(df['followers'], line, color='red', linestyle='--', linewidth=2, label='Trend')

plt.tight_layout()
plt.show()

print(f"Korrelation: {df[['followers', 'engagement_rate']].corr().iloc[0,1]:.3f}")
````

````python
# Sammenlign engagement rate per st√∏rrelses-kategori

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='size_category', y='engagement_rate', palette='Set2')

plt.title('Engagement Rate Distribution by Influencer Size', fontsize=16, fontweight='bold')
plt.xlabel('Influencer St√∏rrelse', fontsize=12)
plt.ylabel('Engagement Rate (%)', fontsize=12)
plt.xticks(rotation=15)


plt.tight_layout()
plt.show()
````

### 6. FORTOLKNING

**Hvad fort√¶ller dataen os?**

1. **Negativ korrelation** mellem f√∏lgere og engagement
2. **Stor variation** inden for hver kategori (ikke perfekt m√∏nster)
3. **Outliers eksisterer** - nogle mega-influencers HAR god engagement

**Hvad betyder det i praksis?**

Hvis I var en virksomhed der skulle v√¶lge en influencer til marketing:

‚ùå **D√•rlig beslutning**: "V√¶lg den med flest f√∏lgere"
- Koster mest, giver mindst engagement

‚úì **God beslutning**: "Analyser engagement rate og relevans"
- Bedre ROI, mere engageret audience

**Dette er v√¶rdien af statistisk t√¶nkning!**
